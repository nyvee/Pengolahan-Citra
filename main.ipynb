```python
import glob
import ultralytics
from ultralytics import YOLO
from IPython.display import Image, display
```

```python
ultralytics.checks()
```

```python
# Load the YOLO model
model = YOLO('yolo11n.pt')

# Train the model
model.train(data='datasets/data.yaml', epochs=100, imgsz=640)
```

```python
model_path = 'runs/detect/train15/weights/best.pt'
data_path = 'datasets/data.yaml'
test_path = 'datasets/test/images'
```

```python
results = YOLO(model_path).val(data=data_path)
print("Validation Results:")
print(results)
```

```python
results = YOLO(model_path).predict(source=test_path, save=True, project='runs/detect', name='predict')
```
