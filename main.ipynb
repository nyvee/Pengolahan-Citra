# **YOLO Object Detection Workflow**

## Setup and Initialization

```python
import glob
import random
import ultralytics
import pandas as pd
from ultralytics import YOLO
from IPython.display import Image, display
```

**Import necessary libraries for object detection**
- **glob**: File path pattern matching
- **ultralytics**: YOLO model framework
- **Image, display**: Jupyter notebook image rendering

## Environment Check

```python
ultralytics.checks()
```

- Verify Ultralytics installation
- Check system compatibility
- Validate dependencies and configurations

## Model Loading

```python
model = YOLO('yolo11n.pt')
```

- Load pre-trained YOLO model
- ***yolo11n.pt***: Nano version of YOLO model
- Lightweight model for faster inference

## Model Training

```python
train_dir = 'runs/detect/train'
model.train(data='datasets/data.yaml', epochs=50, batch=32, imgsz=640, patience=5, project='runs/detect', name='train', exist_ok=True)
```

- Dataset: *datasets/data.yaml*
- Training *epochs*: 50
- batch: 32
- patience: 5
- Image size: 640x640 pixels
- Output directory: *runs/detect/train*
- *exist_ok*=True: Overwrite previous training results

## Visualize Training Results

```python
display(Image(filename=f'{train_dir}/confusion_matrix_normalized.png', width=900))
display(Image(filename=f'{train_dir}/results.png', width=900))
```

```python
df = pd.read_csv('runs/detect/train/results.csv')

print("\nStart of Training Metrics:")
print("Epoch 1:")
print(f"Box Loss: {df.loc[0, 'train/box_loss']:.4f}")
print(f"Classification Loss: {df.loc[0, 'train/cls_loss']:.4f}")
print(f"Precision: {df.loc[0, 'metrics/precision(B)']:.4f}")
print(f"Recall: {df.loc[0, 'metrics/recall(B)']:.4f}")
print(f"mAP50: {df.loc[0, 'metrics/mAP50(B)']:.4f}")

print("\nEnd of Training Metrics:")
last_epoch = len(df) - 1
print(f"Epoch {df.loc[last_epoch, 'epoch']}:")
print(f"Box Loss: {df.loc[last_epoch, 'train/box_loss']:.4f}")
print(f"Classification Loss: {df.loc[last_epoch, 'train/cls_loss']:.4f}")
print(f"Precision: {df.loc[last_epoch, 'metrics/precision(B)']:.4f}")
print(f"Recall: {df.loc[last_epoch, 'metrics/recall(B)']:.4f}")
print(f"mAP50: {df.loc[last_epoch, 'metrics/mAP50(B)']:.4f}")

print("\nPerformance Improvement:")
print(f"Precision Improvement: {df.loc[last_epoch, 'metrics/precision(B)'] - df.loc[0, 'metrics/precision(B)']:.4f}")
print(f"Recall Improvement: {df.loc[last_epoch, 'metrics/recall(B)'] - df.loc[0, 'metrics/recall(B)']:.4f}")
print(f"mAP50 Improvement: {df.loc[last_epoch, 'metrics/mAP50(B)'] - df.loc[0, 'metrics/mAP50(B)']:.4f}")
```

- Display confusion matrix
- Show training performance metrics

## Model Validation

```python
model_path = f'{train_dir}/weights/best.pt'
data_path = 'datasets/data.yaml'
test_path = 'datasets/test/images'
```

```python
results = YOLO(model_path).val(data=data_path, project='runs/detect', name='val', exist_ok=True)
print("Validation Results:")
print(results)
```

- Load best-performing model weights
- Validate model on test dataset
- Generate performance metrics

## Predict

```python
predict_dir = 'runs/detect/predict'
results = YOLO(model_path).predict(source=test_path, save=True, project='runs/detect', name='predict', exist_ok=True)
```

```python
image_paths = glob.glob(f'{predict_dir}/*.jpg')
random_images = random.sample(image_paths, min(12, len(image_paths)))

for image_path in random_images:
    display(Image(filename=image_path, width=300))
    print("\n")
```

- Run object detection on test images
- Save prediction results
- Display first 5 images with detections
- Visually verify model performance
